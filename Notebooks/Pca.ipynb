## Verileri İçeriye Aktar

```python
import pandas as pd 
import numpy as np 

df = pd.read_csv("../Database/Add_Cluster_Feature.csv")
df.head(3)
```

## Preprocessing

Gereksiz sütunları dışarıya çıkar.

```python
df.columns
```

```python
target_col = df[['Level']
                ]
df.drop('Level', axis=1, inplace=True)
```

verileri PCA için standartlaştır.

```python
from sklearn.preprocessing import StandardScaler

sc = StandardScaler()
scaled_data = sc.fit_transform(df)
```

## PCA: Principal Component Analysis

```python
from sklearn.decomposition import PCA

import plotly.express as px
import plotly.graph_objects as go
import matplotlib.pyplot as plt
```

```python
pca = PCA(n_components=3) # 3 Boyutlu bir görselleştirme yapacağımız için `n_component`sayısını 3 olarak seçtim.
pca_result = pca.fit_transform(scaled_data)

pca_df = pd.DataFrame(data=pca_result, columns=['PC1', 'PC2', 'PC3'])
pca_df['Küme'] = df['Küme']

pca_df.head()
```

## PCA Görselleştirme

```python
def get_3d(df, x, y, z, color=None):
    ''''get_3d' işlevi Python'da Plotly Express'i kullanarak bir 3B dağılım grafiği oluşturur.
    
    Parameters
    ----------
    df
        "get_3d" işlevindeki "df" parametresi genellikle 3B dağılım grafiğinde görselleştirmek istediğiniz
    verileri içeren bir DataFrame'dir. Bu DataFrame, çizmek istediğiniz noktaların x, y ve z
    koordinatlarına karşılık gelen sütunlara sahip olmalıdır.
    x
        'get_3d' işlevindeki 'x' parametresi, DataFrame 'df'de 3B dağılım grafiği için x ekseni değerleri
    olarak kullanılacak sütunu temsil eder.
    y
        "get_3d" işlevindeki "y" parametresi, DataFrame "df"de 3B dağılım grafiğinde y ekseni değerleri
    olarak kullanılacak sütunu belirtmek için kullanılır.
    z
        'z' parametresi tipik olarak bir 3B dağılım grafiğinde z ekseni boyunca çizilecek olan veri
    değerlerini temsil eder. Her veri noktasının z ekseni boyunca konumunu belirleyen değişkendir.
    color
        'get_3d' işlevindeki 'color' parametresi, DataFrame 'df'de 3B dağılım grafiğindeki veri noktalarını
    renklendirmek için kullanılacak bir sütunu belirtmek için kullanılır. 'Renk' parametresine bir sütun
    adı ileterek verileri görsel olarak ayırt edebilirsiniz
    
    '''

    fig = px.scatter_3d(
        df,
        x=x,
        y=y,
        z=z,
        color=color
    )

    fig.show()
```

```python
get_3d(pca_df, "PC1", "PC2", "PC3", color="Küme")
```
